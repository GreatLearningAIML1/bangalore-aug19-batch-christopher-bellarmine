```python
ratings_mean_count = pd.DataFrame(df.groupby('ProductId')['Ratings'].mean().sort_values(ascending = False)) 
```

```python
ratings_mean_count
```

```python
ratings_mean_count['rating_counts'] = pd.DataFrame(df.groupby('ProductId')['Ratings'].count())
```

```python
ratings_mean_count
```

```python
mean_rating_dataset = np.mean(ratings_mean_count.Ratings.values)
```

```python
min_votes = 5
```

```python
def weighted_rating(rating_mean_count, min_votes, mean_rating_dataset):
    weighted_rating = list()
    V = rating_mean_count.rating_counts.values
    R = rating_mean_count.Ratings.values
    for v, r in zip(V, R):
        if v < 5:
            weighted_rating.append(0)
        else:
            weighted_rating.append((v/(v+min_votes))*r + (min_votes/(v+min_votes))*mean_rating_dataset)
    return(weighted_rating)
```

```python
ratings_mean_count['weighted_rating'] = weighted_rating(ratings_mean_count, min_votes, mean_rating_dataset)
```

```python
prm=ratings_mean_count.weighted_rating.sort_values(ascending = False).head(10)
```

```python
# Use user_based true/false to switch between user-based or item-based collaborative filtering
algo = KNNWithMeans(k=50, sim_options={'name': 'cosine', 'user_based': True})
algo.fit(trainset)
```

```python
# run the trained model against the testset
test_pred = algo.test(testset)
```

```python
# Evaluate the model to get RMSE
print("Item-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

```python
from collections import defaultdict
```

```python
def get_top_n(predictions, n=10):
    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
top_n = get_top_n(test_pred, n=5)
```
