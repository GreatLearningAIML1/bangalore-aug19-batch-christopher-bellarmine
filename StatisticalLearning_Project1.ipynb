```python
##Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

import pandas as pd
import numpy as np
df = pd.read_csv('Laliga.csv',index_col=0)
df.replace('-',np.NaN,inplace=True)
df.columns=df.iloc[0] ##Setting the attributes row
df.columns
df=df[1:]
df
```

```python
##Print all the teams which have started playing between 1930-1980.

df[['Team','Debut']]
x=df[df['Debut']>str(1929)]  ##x represents the teams who've appeared after 1929
y=x[x['Debut']<str(1980)]    ##y represents the teams who've appeared after 1929 but before 1929 
print(y)
```

```python
##Write a function with name â€œGoal_diff_countâ€ which should
##return all the teams with their Goal Differences.
##Using the same function,find the team which has maximum
##and minimum goal difference.

y=df  ##y is a copy of the same dataframe df (Laliga.cvv)
def Goal_diff_count():
    y['GoalDifferences']=y['GoalsFor'].astype(float)+y['GoalsAgainst'].astype(float)
    maxx=y[y['GoalDifferences']==y['GoalDifferences'].max()]
    minn=y[y['GoalDifferences']==y['GoalDifferences'].min()]
    print("\nThe team with the maximum Goal Differences :\n\n",maxx[['Team','GoalDifferences']])
    print("\nThe team with the minimum Goal Differences :\n\n",minn[['Team','GoalDifferences']])
    return y[['Team','GoalDifferences']]
Goal_diff_count()
```

```python
##Create a new column with name â€œWinning Percentâ€ and append it to
##the data set

y['Percentage of Winning']=100*(y['GamesWon'].astype(float)/y['GamesPlayed'].astype(float))
y.fillna(0)
```

```python
##Group teams based on their â€œBest positionâ€ and print the sum of
##their points for all positions


y['BestPosition']=y['BestPosition'].astype(int)
y['Points'].fillna(0)
y['Points']=y['Points'].astype(float)
y['Points'].groupby([y['BestPosition']]).sum()
```
